```python
%load_ext sql
%sql sqlite://
```

```python
%%sql
CREATE TABLE students (
    id INTEGER,
    name TEXT,
    marks INTEGER
);
```

```python
%%sql
INSERT INTO students VALUES
(1,'Alice',90),
(2,'Bob',80),
(3,'Charlie',95);
```

```python
%%sql
SELECT * FROM students;
```

```python
%%sql

SELECT name, marks
FROM students
WHERE marks > 80;
```

```python
%%sql

SELECT *
FROM students
ORDER BY marks DESC;
```

```python
%%sql

DROP TABLE IF EXISTS students;

CREATE TABLE students (
    id INTEGER,
    name TEXT,
    department TEXT,
    marks INTEGER
);
```

```python
%%sql

INSERT INTO students VALUES
(1,'Alice','CSE',90),
(2,'Bob','ECE',75),
(3,'Charlie','CSE',85),
(4,'David','EEE',80),
(5,'Emma','ECE',95),
(6,'Frank','CSE',70),
(7,'Grace','EEE',88),
(8,'Henry','ECE',92);
```

```python
%%sql
SELECT department, AVG(marks)
FROM students
GROUP BY department;
```

```python
%%sql
SELECT department, COUNT(*)
FROM students
GROUP BY department
HAVING COUNT(*) > 2;
```

```python
%%sql

CREATE TABLE departments (
    department TEXT,
    hod TEXT
);

INSERT INTO departments VALUES
('CSE','Dr. Rao'),
('ECE','Dr. Sharma'),
('EEE','Dr. Kumar');
```

```python
%%sql
SELECT s.name, s.department, d.hod
FROM students s
JOIN departments d
ON s.department = d.department;
```
