```python
#practice1
age = 25
print(f"Hello, I am {age} years old.")
```

```python
#practice2 - operators
a = 10
b = 3
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
```

```python
#practice 3 - checking even or odd
num = 7
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
#practice4 - loop
num = 5
for i in range(1, 11):
    print(f"{num} x {i} = {num*i}")
```

```python
#practice5 - list
fruits = ["apple", "banana", "mango", "orange"]
fruits.append("grapes")
print("Fruit list:", fruits)
print("First fruit:", fruits[0])
print("Total fruits:", len(fruits))
```

```python
#practice5 - dictionary
student = {"name": "Priya", "marks": 89, "grade": "A"}
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
#practice6 - manipulating string
text = "Python is Fun"
print("Uppercase:", text.upper())
print("Lowercase:", text.lower())
print("Reversed:", text[::-1])
print("Word count:", len(text.split()))
```
